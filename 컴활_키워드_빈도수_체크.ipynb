```python
from konlpy.tag import Okt
from collections import Counter
import csv
```

```python
f = open("/content/drive/MyDrive/2021 창업동아리/컴활.txt",'r',encoding='utf-8')
data = f.read()
f.close()
```

```python
okt=Okt()
noun = okt.nouns(data)
count = Counter(noun)
```

```python
noun_list = count.most_common(10000)
for v in noun_list:
  print(v)
```

```python
with open("noun_list.txt",'w',encoding='utf-8') as f:
  for v in noun_list:
    f.write("".join(map(str,v)))
    f.write("\n")
```

```python
with open("noun_list.csv","w",newline='',encoding='euc-kr') as f:
  csvw = csv.writer(f)
  for v in noun_list:
    csvw.writerow(v)
```

#불용어 처리

```python
!pip3 show nltk
```

```python
import nltk
from nltk.corpus import stopwords 
from nltk.tokenize import word_tokenize 
```

```python
nltk.download('punkt')
```

```python
f = open("/content/drive/MyDrive/2021 창업동아리/koreanStopwords.txt", 'r')
lines = f.readlines()
stopwords = []
for line in lines:
    line = line.replace('\n', '')
    stopwords.append(line)
f.close()
```

```python
noun_list_stopwords=[]
for w in noun: 
    if w not in stopwords: 
        noun_list_stopwords.append(w) 

print(noun,'\n') 
print(noun_list_stopwords)
```

```python
stop_count = Counter(noun_list_stopwords)
```

```python
stop_noun_list = stop_count.most_common(10000)
for v in stop_noun_list:
  print(v)
```

```python
with open("stop_noun_list.txt",'w',encoding='utf-8') as f:
  for v in stop_noun_list:
    f.write("".join(map(str,v)))
    f.write("\n")
```

```python
with open("stop_noun_list.csv","w",newline='',encoding='euc-kr') as f:
  csvw = csv.writer(f)
  for v in stop_noun_list:
    csvw.writerow(v)
```

```python
wc.generate_from_frequencies(dict(stop_noun_list))
wc.to_file('컴활기출문제 워드클라우드_불용어처리.png')
```

```python
with open("stopwords.txt",'w',encoding='utf-8') as f:
  for v in stopwords:
    f.write("".join(map(str,v)))
    f.write("\n")
```

```python
with open("stopwords.csv","w",newline='',encoding='euc-kr') as f:
  csvw = csv.writer(f)
  for v in stopwords:
    csvw.writerow(v)
```
