```python
import re
from bs4 import BeautifulSoup
from collections import Counter
from konlpy.tag import *

kkma = Kkma()
hannanum = Hannanum()
komoran = Komoran()
okt = Okt()
```

```python
data_txt = open('chapter_1.txt', 'r',  encoding='UTF8')
txt_read = data_txt.readlines()
txt_read = list(map(lambda x: x.strip(), txt_read))
txt_read = list(map(lambda x: x.replace("\xa0", ""), txt_read))
txt_read = list(map(lambda x: x.replace("\n", ""), txt_read))
txt_read = list(map(lambda x: x.replace(" ", ""), txt_read)) # 띄어쓰기 지우기
txt_read = list(map(lambda x: re.sub(r'[①②③④]', '', x), txt_read)) # 선택지 번호 지우기
txt_read = list(map(lambda x: re.sub(r'^[0-9]+\.', '', x), txt_read))
txt_read_toString = "".join(txt_read)
```

```python
token = okt.nouns(txt_read_toString)
filter_word = ['인터넷', '정보', '다음', '설정', '연결', '중', '컴퓨터', '대한', 
               '작업', '설명','기능', '경우', '사용', '관련', '가장', '위해', 
               '하나', '한글', '해당', '특징', '구성', '기법', 
               '방법', '방식', '한글', '거나', '통해', '을사', '한비', 
               '실행', '파일', '장치', '해하', '부분', '내용',
               '여러', '로사', '장하', '로만', '및', '종자',
              '이용', '설치', '수행', '기술', '서비스', '시스템']
token_process = list(filter(lambda x : len(x) > 1 and x not in filter_word, token))
counts = Counter(token_process)
token_to_tags = counts.most_common(10000)
```

```python
result = []
for i in range(len(txt_read)):
    token_list = okt.pos(txt_read[i], stem=True, norm=True)
    temp = []
    for word in token_list:
        if ((word[1] in ["Noun"]) and (word[0] not in filter_word) and (len(word[0]) > 2)):
            temp.append((word[0]))
            
    if temp:
        result.append(temp)
```
